###LINIEAR REGRESSION

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

X = np.random.rand(100, 1)
y = 2 * X + 1 + 0.1 * np.random.rand(100, 1)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

model.compile(optimizer='sgd', loss='mean_squared_error')

costs = []
weights = []

history = model.fit(X, y, epochs=100, verbose=0)

for loss in history.history['loss']:
    costs.append(loss)
    weights.append(model.get_weights())

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(costs)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs. Epochs (Regresi Linier)')

weights = np.array(weights)
plt.subplot(1, 2, 2)
for i in range(weights.shape[1]):
    plt.plot(weights[:, i, 0], label=f'Weight {i}')
plt.xlabel('Epochs')
plt.ylabel('Weight Value')
plt.title('Weight Updates vs. Epochs (Regresi Linier)')
plt.legend()
plt.tight_layout()
plt.show()
```

###Logistic regression

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

X = np.random.rand(100, 2)
y = (X[:, 0] + X[:, 1] > 1).astype(int)

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

model.compile(optimizer='sgd', loss='binary_crossentropy')

costs = []
weights = []

history = model.fit(X, y, epochs=100, verbose=0)

for loss in history.history['loss']:
    costs.append(loss)
    weights.append(model.get_weights()[0])

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(costs)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs. Epochs (Regresi Logistik)')

weights = np.array(weights)
plt.subplot(1, 2, 2)
for i in range(weights.shape[1]):
    plt.plot(weights[:, i, 0], label=f'Weight {i}')
plt.xlabel('Epochs')
plt.ylabel('Weight Value')
plt.title('Weight Updates vs. Epochs (Regresi Logistik)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import tensorflow as tf

#data pelanggan contoh
data_pelanggan = np.array([
    [25, 300, 0],  # Usia 25 tahun, Pengeluaran 300, Tidak memiliki riwayat pembelian sebelumnya
    [30, 400, 1],  # Usia 30 tahun, Pengeluaran 400, Memiliki riwayat pembelian sebelumnya
    [35, 200, 0],  # Usia 35 tahun, Pengeluaran 200, Tidak memiliki riwayat pembelian sebelumnya
    [22, 150, 0],  # Usia 22 tahun, Pengeluaran 150, Tidak memiliki riwayat pembelian sebelumnya
])

X = data_pelanggan[:, :2]
Y = data_pelanggan[:, 2]

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=25, activation='sigmoid'),
    tf.keras.layers.Dense(units=15, activation='sigmoid'),
    tf.keras.layers.Dense(units=1, activation='linear')
])

model.compile(optimizer='adam', loss=BinaryCrossentropy(from_logits=True))
model.fit(X, Y, epochs=100, verbose=1)
pelanggan_baru = np.array([
    [28, 350],
])

prediksi = model.predict(pelanggan_baru)
probabilitas = tf.nn.sigmoid(prediksi)

for i in range(len(pelanggan_baru)):
    print(f'Pelanggan baru {i + 1}:')
    print(f'Usia: {pelanggan_baru[i][0]} tahun')
    print(f'Pengeluaran: {pelanggan_baru[i][1]}')
    print(f'Probabilitas Membeli: {probabilitas[i][0]}')
    print()
```
